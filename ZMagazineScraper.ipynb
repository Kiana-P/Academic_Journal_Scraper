```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# function to sort body text
def clean_body(text: str) -> dict:
    data = {}
    
    # every article ends with a Z followed by a \n character so we split it from the byline text that follows
    split_string = text.split('Z\n', 1)
    substring = split_string[0]
    byline = split_string[1]
    byline = byline.replace('\n', ' ').replace('\xa0', ' ')
    
    # every article starts with text about the photo that we can exclude
    split_string = substring.split('\n\xa0', 1)
    text = split_string[1]
    text = text.replace('\n', ' ').replace('\xa0', ' ')
    
    # store fields in dictionary
    data['text'] = text
    data['byline'] = byline
    
    return data
```

```python
# function to sort header text
def clean_header(text: str) -> dict:
    data = {}
    
    # split text by \n characters
    data['header'] = text
        
    return data
```

```python
# function to get data from url
def scrape_data(url: str, post_list: pd.DataFrame()):
    
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    #gets text and byline
    for data in soup.find_all('div', class_='entry-content'):
        text = data.get_text()
        
    text_data = clean_body(text)
    
    #gets title, author, date
    for data in soup.find_all('div', class_='post-header'):
        text = data.get_text()
        
    post_data = clean_header(text)
    post_data.update(text_data)
    
    post_list.append(post_data)
```
